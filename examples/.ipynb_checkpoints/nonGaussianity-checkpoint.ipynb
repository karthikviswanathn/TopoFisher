```python
import imph.inputs.noisy_ring
from imph.inputs.noisy_ring import CircleSimulator
import imph.filtrations.numpy.alphaDTML
from imph.filtrations.numpy.alphaDTML import findST, findCC
import imph.vectorizations.numpy.TOPK
from imph.vectorizations.numpy.TOPK import TopKBinned, TopKBD, TopKBDP
from imph.vectorizations.numpy import utils
import tensorflow as tf
import numpy as np
import tqdm
from tqdm import tqdm
from imph.fisher.Fisher import fisherMOPED, fisherFromVecs
from imph.fisher.nnutils import trainAndValidate
import powerbox as pbox
import matplotlib.pyplot as plt
```

```python
physical = False
N = 32
A, B, f = 1.1, 1.7, 0.04
theta = [A, B, f]
deltaA, deltaB, deltaF = 0.01, 0.01, 0.0004
delta_theta = tf.constant([deltaA, deltaB, deltaF])

def simulate(thet, n_sims, seed = None):
    a, b, f = thet
    if(seed != None) : np.random.seed(seed)
    seeds = np.random.randint(1e7, size = (n_sims))
    lis = []
    for i in range(n_sims):
        pb = pbox.PowerBox(
            N=N,                     # Number of grid-points in the box
            dim=2,                     # 2D box
            pk = lambda k: a*k**(-b), # The power-spectrum
            boxlength = N,           # Size of the box (sets the units of k in pk)
            vol_normalised_power=True, 
            ensure_physical = physical,
            seed = seeds[i]
        )
        z = pb.delta_x() + f * (pb.delta_x()**2 - np.mean(pb.delta_x())**2)
        lis.append(z)
    return np.array(lis)
```

```python
n_s = 4000; n_d = 4000; dim_list = [0, 1]; seed = np.random.randint(1e4); 
print("seed = ", seed)
thetaLis = [[A + deltaA, B, f], [A, B + deltaB, f], [A, B, f + deltaF]] 

ptsCov = simulate(theta, n_s)
pts = simulate(theta, n_d, seed = seed)
pts_pert = [simulate(thet, n_d, seed) for thet in thetaLis]
all_pts = [ptsCov, pts, *pts_pert]
```

## Filtration 

```python
all_st = []; 
for ptclouds in all_pts:
    st_list = [findCC(sim) for sim in tqdm(ptclouds)]
    all_st.append(st_list)
```

## Vectorization

### TopK Binned

```python
veclayer = TopKBinned(topks = [65, 100], num_bins = [10, 15])
all_vecs = []; 
for st_list in all_st:
    vecs = [veclayer(item) for item in st_list]
    all_vecs.append(vecs)
all_vecs_binned = np.array(all_vecs)
```

### TopK sorted

```python
veclayer = TopKBD(topks = [65, 100])
all_vecs = []; 
for st_list in all_st:
    vecs = [veclayer(item) for item in st_list]
    all_vecs.append(vecs)
all_vecs_bd = np.array(all_vecs)
```

## Fisher Analysis

### Thoeretical FI

```python
def fisher(θ, k, N):
    A, B = θ[:2]
    k = k[0:N//2, 0:N].flatten()
    Pk = A*k**(-B)

    Cinv = np.diag(1. / Pk)
    C_A =  np.diag(k ** -B)
    C_B =  np.diag(- Pk * np.log(k))

    F_AA = 0.5 * np.trace((C_A @ Cinv @ C_A @ Cinv))
    F_AB = 0.5 * np.trace((C_A @ Cinv @ C_B @ Cinv))
    F_BA = 0.5 * np.trace((C_B @ Cinv @ C_A @ Cinv))
    F_BB = 0.5 * np.trace((C_B @ Cinv @ C_B @ Cinv))

    return np.array([[F_AA, F_AB], [F_BA, F_BB]])

pb = pbox.PowerBox(
    N = N,                     # Number of grid-points in the box
    dim = 2,                     # 2D box
    pk = lambda k: A*k**(-B), # The power-spectrum
    boxlength = N,           # Size of the box (sets the units of k in pk)
    vol_normalised_power=True, 
    ensure_physical = physical,
)

TFM = fisher(theta, pb.k(), N)
```

### Fisher Matrix calculation - binned + MOPED

```python
fish = fisherMOPED(all_vecs_bd, delta_theta)
fisherMatrix = fish.FM.numpy()
print("The fisher matrix for this pipeline is ", fish.FMnp)
print("The theoretical fisher matrix for (A, B) is ", np.round(TFM.flatten()))

print("\nThe fisher information using MOPED is ", np.log(np.linalg.det(fisherMatrix)))
```

### Fisher Matrix from sorted bd + compressing using neural networks 

```python
inp_vecs = all_vecs_bd
div = (np.array([0.5,  0.25, 0.25]) * inp_vecs.shape[1]).astype(int)
div = np.cumsum(div)
train_vecs, valid_vecs, test_vecs = inp_vecs[:, :div[0]], inp_vecs[:, div[0]:div[1]], inp_vecs[:, div[1]:div[2]]

hidden_sz = inp_vecs.shape[-1]//2
comp = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_sz, activation = tf.keras.layers.LeakyReLU(alpha=0.01)),
    tf.keras.layers.Dense(len(theta))])
```

```python
comp_final = trainAndValidate(delta_theta, train_vecs, valid_vecs, test_vecs, comp, tfm = None, num_epochs = 75, batch_size = 256)
```
