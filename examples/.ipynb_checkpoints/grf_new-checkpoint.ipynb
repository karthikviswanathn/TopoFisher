## Top-k plot

```python
importlib.reload(imph.vectorizations.numpy.TOPK)
from imph.vectorizations.numpy.TOPK import TOPKLayers
```

```python
topks = [item.topk for item in topk_vec_layer.veclayers]
topks
```

```python
def getFish(topks_new):
    bins = [np.maximum(2, topks_new[0]//10), np.maximum(2, topks_new[1]//10)]  
    # print(topks_new)
    # print("bins = ", bins)
    fish = grf_pipeline.vectorize_and_fisher(\
             TOPKLayers(bdp_type = "bp", topks = topks_new, is_binned = True, num_bins = [10, 10], hom_dim_list = [0, 1],\
                        quantile = 0.95),\
            MopedLayer())
    fishMatrix = fish.FMnp
    return fishMatrix[0], fishMatrix[-1], fish.lnDetF
```

```python
k0r = np.linspace(10, topks[0], 10, dtype = np.int32)
k1r = np.linspace(10, topks[1], 10, dtype = np.int32)
fisher_dict = {}
for idx1 in tqdm(k1r):
    for idx0 in k0r:
        # print(idx0, idx1)
        inds = (idx0, idx1)
        fisher_dict[inds] = getFish(inds)            
```

```python
graph_list = []
def getTopK(tp):
    # print(tval)
    dic = {}
    for idx1 in k1r:
        lis = []
        for idx0 in k0r:
            inds = (idx0, idx1)
            lis.append(fisher_dict[inds][tp])
        dic[idx1] = lis
    return dic

for idx in range(3): 
    graph_list.append(getTopK(idx))
```

```python
def plotTopK(dic, ax, title, tval):
    for idx in k1r: 
        ax.plot(k0r, dic[idx], label = "k1 = " + str(idx))
        
    
    ax.set_xlabel('k0')
    ax.set_title(title)
    ax.set_ylabel('FI')

    ax.axhline(tval, linestyle = 'dashed', c = 'red', label = "theoretical", alpha = 0.7)

tfm = grf_pipeline.benchmarks['TFM']
log_tfi = np.log(np.linalg.det(tfm)) 
tfi_lis = [tfm[0,0], tfm[-1,-1], log_tfi]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))   
names = ['FI(A)', 'FI(B)', 'log_tfi']

for idx in range(3): 
    plotTopK(graph_list[idx], axes[idx], names[idx], tfi_lis[idx])
    
# handles, labels = axes[0].get_legend_handles_labels()
# fig.legend(handles, labels, loc = 'lower right')
fig.suptitle("Top-k graphs for A, B = " + str(np.round(grf_pipeline.theta_fid.numpy(), 2)))
plt.tight_layout()
# plt.show()
plt.savefig('plots/grf/topk_graphs.png' , bbox_inches='tight')
```
