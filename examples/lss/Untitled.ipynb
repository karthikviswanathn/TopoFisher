```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Plot the Leaky ReLU activation function for different alpha values using TensorFlow's LeakyReLU layer
alpha_values = [0.1, 0.2, 0.5]

plt.figure(figsize=(8, 6))

for alpha in alpha_values:
    # Create a LeakyReLU layer with the specified alpha
    leaky_relu_layer = tf.keras.layers.LeakyReLU(alpha=alpha)
    
    # Apply the Leaky ReLU activation function
    y_values = leaky_relu_layer(x_values)
    
    plt.plot(x_values, y_values, label=f'Leaky ReLU (alpha={alpha})')

# Plot settings
plt.title('Leaky ReLU Activation Function (TensorFlow)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Apply Leaky ReLU with a small alpha
alpha = 0.05
leaky_relu_layer = tf.keras.layers.LeakyReLU(alpha=alpha)
y_values_leaky_relu = leaky_relu_layer(x_values)

# Standard ReLU
relu_layer = tf.keras.layers.ReLU()
y_values_relu = relu_layer(x_values)

# Plot the Leaky ReLU and ReLU activation functions
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_leaky_relu, label=f'Leaky ReLU (alpha={alpha})', linestyle='--')
plt.plot(x_values, y_values_relu, label='ReLU')
plt.title('Leaky ReLU vs. ReLU Activation Function (TensorFlow)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
